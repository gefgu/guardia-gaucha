```python
import pydeck as pdk

# Define the map layer
layer = pdk.Layer(
    "ScatterplotLayer",
    data=[{"position": [-122.4, 37.8], "size": 100}],
    get_position="position",
    get_radius="size",
    get_color=[255, 0, 0],
)

# Define the view state
view_state = pdk.ViewState(
    latitude=37.8,
    longitude=-122.4,
    zoom=14,
    bearing=0,
    pitch=0,
)

# Create the deck.gl map
r = pdk.Deck(layers=[layer], initial_view_state=view_state)

# Render the map
r.to_html("pydeck_map.html", notebook_display=False)

html_content = """
<!DOCTYPE html>
<html>
<head>
</head>
<body>
  <div id="map-container">
    <iframe src="pydeck_map.html" style="width:100%; height:100%; border:none;"></iframe>
    <div id="overlay">I</div>
  </div>
</body>
</html>
"""

# Save the HTML content to a file
with open("overlay_test.html", "w") as file:
    file.write(html_content)
```
